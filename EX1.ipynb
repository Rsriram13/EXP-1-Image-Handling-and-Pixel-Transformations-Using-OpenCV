```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
apollo = cv2.imread("Apollo-11-launch.jpg")
apollo_rgb = cv2.cvtColor(apollo, cv2.COLOR_BGR2RGB)
```

```python
text = "Apollo 11 Saturn V Launch, July 16, 1969"
font_face = cv2.FONT_HERSHEY_PLAIN

h, w, c = apollo.shape
text_size = cv2.getTextSize(text, font_face, 2, 2)[0]
text_x = (w - text_size[0]) // 2
text_y = h - 30

cv2.putText(apollo, text, (text_x, text_y),
            font_face, 2, (255, 255, 255), 2)
```

```python
# Magenta Rectangle (Correct Position)
rect_color = (255, 0, 255)  # Magenta (BGR)

# Correct bounding box coordinates
cv2.rectangle(apollo, (520, 90), (720, 640), rect_color, 3)
```

```python
apollo_rgb = cv2.cvtColor(apollo, cv2.COLOR_BGR2RGB)

plt.imshow(apollo_rgb)
plt.title("Annotated Apollo Launch Image")
plt.axis("off")
plt.show()
```

```python
img = cv2.imread("boy.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
matrix_ones = np.ones(img.shape, dtype="float64") * 50
```

```python
img_float = img.astype(np.float64)
img_brighter = cv2.add(img_float, matrix_ones)
img_darker = cv2.subtract(img_float, matrix_ones)

img_brighter = np.clip(img_brighter, 0, 255).astype(np.uint8)
img_darker = np.clip(img_darker, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img_darker.astype("uint8"), cv2.COLOR_BGR2RGB))
plt.title("Darker")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img_brighter.astype("uint8"), cv2.COLOR_BGR2RGB))
plt.title("Brighter")
plt.axis("off")

plt.show()
```

```python
img_higher1 = cv2.convertScaleAbs(img, alpha=1.1, beta=0)
img_higher2 = cv2.convertScaleAbs(img, alpha=1.2, beta=0)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img_higher1, cv2.COLOR_BGR2RGB))
plt.title("Contrast =1.1")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img_higher2, cv2.COLOR_BGR2RGB))
plt.title("Contrast =1.2")
plt.axis("off")

plt.show()
```

```python
b, g, r = cv2.split(img)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(b, cmap="gray")
plt.title("Blue Channel")

plt.subplot(1,3,2)
plt.imshow(g, cmap="gray")
plt.title("Green Channel")

plt.subplot(1,3,3)
plt.imshow(r, cmap="gray")
plt.title("Red Channel")

plt.show()
```

```python
merged_rgb = cv2.merge([r, g, b])

plt.imshow(merged_rgb)
plt.title("Merged RGB Image")
plt.axis("off")
plt.show()
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(h, cmap="gray")
plt.title("Hue Channel")

plt.subplot(1,3,2)
plt.imshow(s, cmap="gray")
plt.title("Saturation Channel")

plt.subplot(1,3,3)
plt.imshow(v, cmap="gray")
plt.title("Value Channel")

plt.show()
```

```python
merged_hsv = cv2.merge([h, s, v])

plt.imshow(merged_hsv)
plt.title("Merged HSV Image")
plt.axis("off")
plt.show()
```
